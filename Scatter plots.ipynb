```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,4,6,9,11])
y = np.array([55,60,50,64,70,56,53,75,80])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige"])

plt.scatter(x, y, c=colors)

font1 = {'family':'serif','color':'green','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Employee Details:",fontdict = font1  )
plt.xlabel("Experience",fontdict = font2)
plt.ylabel("Age",fontdict = font2)
plt.show()
```
